```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data=pd.read_csv('imports.csv')
data.head()
```

```
data.isnull().sum()
```

```
quant = []
cat = []

for i in data.columns:
  if len(data[i].unique())>15:
    quant.append(i)
  else:
    cat.append(i)
```

```
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='mean')
data[quant]=imputer.fit_transform(data[quant])
```

```
data
```

```
quant
```

```
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='most_frequent')
data[cat]=imputer.fit_transform(data[cat])
```

```
cat
```

```
sns.boxplot(cat)
```

```
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cat:
  data[i]=le.fit_transform(data[i])
```

```
q1=data[quant].quantile(0.25)
q3=data[quant].quantile(0.75)
iqr=q3-q1
lower_bound=q1-(1.5*iqr)
upper_bound=q3+(1.5*iqr)
data_final=data[~((data[quant]<(lower_bound))|(data[quant]>(upper_bound))).any(axis=1)]
```

```
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
data[quant]=ss.fit_transform(data[quant])
```

```
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
data[quant]=mm.fit_transform(data[quant])
```

```
corr=data.corr()
labeled_corr=corr[['price']].sort_values(by='price',ascending=False)
sns.heatmap(labeled_corr,annot=True,cmap='coolwarm')
```
